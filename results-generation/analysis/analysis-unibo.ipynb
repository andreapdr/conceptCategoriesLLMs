# Magnitude Correlations

```python
import pandas as pd
```

```python
corrs = pd.read_csv("magnitudes_corr.csv")
corrs = corrs.drop(columns="temp")
corrs = corrs.drop(columns="human")
```

```python
t00 = corrs.iloc[0]
t05 = corrs.iloc[1]
t10 = corrs.iloc[2]
```

```python
t00
```

```python
import matplotlib.pyplot as plt

plotstyle = {
    "edgecolor": "black",
    "linewidth": 0.5
}

fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
axs = axs.ravel()

axs[0].bar(x=t00.index, height=t00.values, **plotstyle)
axs[1].bar(x=t05.index, height=t05.values, **plotstyle)
axs[2].bar(x=t10.index, height=t10.values, **plotstyle)

axs[0].set_xticklabels([model_name.split("-simple")[0] for model_name in t00.index], rotation=90)
axs[0].set_title("temp 0.0")

axs[1].set_xticklabels([model_name.split("-simple")[0] for model_name in t00.index], rotation=90)
axs[1].set_title("temp 0.5")

axs[2].set_xticklabels([model_name.split("-simple")[0] for model_name in t00.index], rotation=90)
axs[2].set_title("temp 1.0")

plt.suptitle("Magnitude correlations human/model")

fig.show()
```
